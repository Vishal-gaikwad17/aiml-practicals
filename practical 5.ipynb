```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Create a directed graph
G = nx.DiGraph()

# 2. Define Nodes (Concepts)
G.add_nodes_from([
    "Dog", "Cat", "Animal", "Mammal",
    "CanFly", "HasFur", "Meows", "Barks"
])

# 3. Define Relationships (Edges)
# 'is-a' relationships
G.add_edge("Dog", "Mammal", label="is-a")
G.add_edge("Cat", "Mammal", label="is-a")
G.add_edge("Mammal", "Animal", label="is-a")

# 'can' / 'has' relationships
G.add_edge("Dog", "Barks", label="can")
G.add_edge("Dog", "HasFur", label="has")
G.add_edge("Cat", "Meows", label="can")
G.add_edge("Cat", "HasFur", label="has")
G.add_edge("Animal", "CanFly", label="cannot")

# 4. Inference
is_dog_an_animal = nx.has_path(G, "Dog", "Animal")
print("Is a Dog an Animal?", is_dog_an_animal)

# 5. Visualization
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 7))

nx.draw_networkx_nodes(G, pos, node_color="lightblue", node_size=2000)
nx.draw_networkx_labels(G, pos, font_size=11, font_weight="bold")
nx.draw_networkx_edges(G, pos, edge_color="black", arrows=True)

# Edge labels
edge_labels = nx.get_edge_attributes(G, "label")
nx.draw_networkx_edge_labels(G, pos, edge_labels= edge_labels)

plt.title("Simple Semantic Network for Knowledge Representation")
plt.axis("off")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes (concepts)
G.add_node("Animal")
G.add_node("Mammal")
G.add_node("Dog")
G.add_node("Bird")
G.add_node("Tail")
G.add_node("Fly")

# Add edges (relationships)
G.add_edge("Dog", "Mammal", relation="is-a")
G.add_edge("Mammal", "Animal", relation="is-a")
G.add_edge("Dog", "Tail", relation="has")
G.add_edge("Bird", "Animal", relation="is-a")
G.add_edge("Bird", "Fly", relation="can")

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10)

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```
