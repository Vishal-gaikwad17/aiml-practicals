```python
import itertools
def is_tautology(expr):
    variables = ['p','q']
    for p, q in itertools.product((True,False), repeat = 2):
        if not eval(expr):
            print("Not a Tautology")
            return False 

        print("It is tautology")
        return True

expr = input("Enter Expressions using p, q, and, or , not")
is_tautology(expr)
```
